# Sort a dataframe

```python
# sort by one column
df_exams.sort_values(by='math score')
```

```python
# sort descending by one column
df_exams.sort_values(by='math score',ascending=False)
```

```python
# sort descending by multiple column
# order of column name give priority
# in this case math score > reading score 
df_exams.sort_values(['math score','reading score'],ascending=False)
```

```python
df_exams.sort_values(['math score','reading score'],ascending=False,
                    inplace=True)
df_exams
```

```python
# sort descending with a key function
df_exams.sort_values('race/ethnicity',ascending=True,
                    key=lambda col:col.str.lower())
```

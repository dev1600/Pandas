```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df=pd.DataFrame(people)
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
df
```

```python
# Remove NaN
# the subset takes list of column that should have have value
df.dropna(axis='index',how='all',subset=['email'])
```

```python
#! Here it drops the index if that index has  
# both email and last as NaN
df.dropna(axis='index',how='all',subset=['email','last'])
```

```python
# replaces Nan with other data
# useful when working numeric data 
# and assign them default value
df.fillna('Missing')
```

```python
# convert age column to float from str
# convert age to float since NaN is float type
df['age']=df['age'].astype(float)
```
